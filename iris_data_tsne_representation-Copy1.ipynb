```python
from tsne import bh_sne
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data

y = iris.target

X_2d = bh_sne(X)

plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tsne_data=pd.DataFrame(data=X_2d,columns=["first","second"])
```

```python
tsne_data["label"]=iris.target
```

```python
sns.FacetGrid(tsne_data,hue="label",height=10).map(plt.scatter,"first","second")
```
