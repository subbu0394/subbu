```python
from sklearn.manifold import TSNE
```

```python
model=TSNE(n_components=2,random_state=0,verbose=1)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\naruk\OneDrive\Desktop\ammu\mnist_test.csv")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sample_data=sc.fit_transform(data)
```

```python
tsne_data=model.fit_transform(sample_data)
```

```python
tsne_data=np.vstack((tsne_data.T,data.label)).T
```

```python
tsne_df=pd.DataFrame(data=tsne_data,columns=("first","second","label"))
```

```python
import matplotlib.pyplot as plt
```

```python
sns.FacetGrid(tsne_df,hue="label",size=40).map(plt.scatter,'first','second')
```
