```python
data["Division.Name"].unique()
```

```python
div=data.pop("Division.Name")
```

```python
data['general']=(div=="General")*1.0
data['petite']=(div=="General Petite")*1.0
```

```python
dep=data.pop("Department.Name")
```

```python
dep.unique()
```

```python
data["dress"]=(dep=="Dresses")*1.0
data["jacket"]=(dep=="Jackets")*1.0
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(data)
```

```python
y=kmeans.fit_predict(data)
```

```python
data["cluster"]=y
```

```python
df=data.groupby("cluster")
df1=df.get_group(0)
```

```python
df2=df.get_group(1)
#df3=df.get_group(2)
#df4=df.get_group(3)
```

```python
plt.plot(df1["Clothing.ID"],(df1.Age),"go")
plt.plot(df2["Clothing.ID"],(df2.Age),"yo")
#plt.plot(df3["Clothing.ID"],(df3.index),"bo")
#plt.plot(df4["Clothing.ID"],(df4.index),"ro")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='green',marker='*',label="centroid")
```
