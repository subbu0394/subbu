```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tkinter import *
tf.compat.v1.disable_eager_execution()
```

```python
a=input("enter sequence")
l=a.split(" ")
for i in range(0,len(l)):
    l[i]=int(l[i])
li=np.asarray(l)
m=tf.Variable([1.0],tf.float32)
c=tf.Variable([1.0],tf.float32)
x=tf.compat.v1.placeholder(tf.float32)
y=tf.compat.v1.placeholder(tf.float32)
l=m*x+c
init=tf.compat.v1.initializers.global_variables()
sess=tf.compat.v1.Session()
sess.run(init)
s=tf.square(l-y)
loss=tf.reduce_sum(s)
opt=tf.compat.v1.train.GradientDescentOptimizer(0.01)
train=opt.minimize(loss)
for i in range(1000):
    sess.run(train,{x:[1,2,3,4,5],y:li})
m,c=sess.run([m,c])
v=m*6+c
print(v)
v=v.tolist()
a=float(v[0])
print(round(a))


```

```python
window=Tk()
window.title("veer")
l1=Label(window,text="enter sequence ")
l1.pack()
n1=Entry(window,width=10)
n1.pack()
l3=Label(window,text="")
l3.pack()

def click1():
    a=n1.get()
    l=a.split(" ")

    l[0]=int(l[0],base=10)
    l[1]=int(l[1],base=10)
    l[2]=int(l[2],base=10)
    l[3]=int(l[3],base=10)
    l[4]=int(l[4],base=10)
    li=np.asarray(l)
    m=tf.Variable([1.0],tf.float32)
    c=tf.Variable([1.0],tf.float32)
    x=tf.placeholder(tf.float32)
    y=tf.placeholder(tf.float32)
    l=m*x+c
    init=tf.global_variables_initializer()
    sess=tf.Session()
    sess.run(init)
    s=tf.square(l-y)
    loss=tf.reduce_sum(s)
    opt=tf.train.GradientDescentOptimizer(0.01)
    train=opt.minimize(loss)
    for i in range(5000):
        sess.run(train,{x:[1,2,3,4,5],y:li})
    m,c=sess.run([m,c])
    v=m*6+c
    v=v.tolist()
    am=float(v[0])
    veer=round(am)
    l3.configure(text="your answer is ="+str(veer))
bt1=Button(window,text="next number in the sequence is",command=click1).pack()
window.geometry("600x600")
window.mainloop()
```
